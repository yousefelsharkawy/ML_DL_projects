```python
#mlpformulticlassclassification
from numpy import argmax 
from pandas import read_csv 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
import numpy as np
```

# multiclass classification

```python
#load the dataset
path='https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv'
df=read_csv(path,header=None)
```

```python
df
```

```python
df[4].unique()
```

```python
df.values
```

```python
#split into input and output columns 
X,y=df.values[:,:-1],df.values[:,-1]
```

```python
X.shape,y.shape
```

```python
#ensure all data are floating point values 
X=X.astype('float32')
```

```python
#encode strings to integer 
y=LabelEncoder().fit_transform(y)
```

```python
y
```

```python
#split into train and test data sets
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.33)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
# define the model 
n_dimension = x_test.shape[1]
model = Sequential([
    Dense(10,activation='relu',input_shape=((n_dimension,))),
    Dense(8,activation='relu'),
    Dense(3,activation='softmax')
])
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam() , loss="sparse_categorical_crossentropy")
```

```python
model.fit(x_train,y_train,epochs=500)
```

## evaluate the model

```python
y_prid = model.predict(x_test)
```

```python
y_prid.shape
```

```python
y_prid = np.argmax(y_prid,axis=1)
```

```python
y_prid = y_prid.astype('int32')
```

```python
y_prid,y_test
```

```python
(y_prid == y_test)
```

```python
np.where(y_prid != y_test)[0][0]
```

```python
y_prid[32],y_test[32]
```

```python
(y_prid == y_test).sum() / len(y_test)
```

```python
model.evaluate(x_test,y_test)
```

# regression model

```python
# load the dataset 
path='https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv'
df=read_csv(path,header=None)
```

```python
df
```

```python
# split the inputs and the label 
x = df.values[:,:-1]; y = df.values[:,-1]
```

```python
x.shape,y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

## define the model 

```python
model = Sequential([
    Dense(30,activation = 'relu'),
    Dense(20,activation = 'relu'),
    Dense(8,activation = 'relu'),
    Dense(1,activation = 'linear')
])
```

```python
model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.001),loss='mse')
```

```python
model.fit(X_train,y_train,epochs=300)
```

```python
error = model.evaluate(X_test,y_test)
```

```python
print(f"MSE = {error}, RMSE = {np.sqrt(error)}")
```

```python
y_prid = model.predict(X_test)
y_prid
```

```python
y_test
```

```python
abs(y_prid - y_test).mean()
```
